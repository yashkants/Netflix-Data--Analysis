## Insight 1:How many movies or shows are present in this dataset?

```python
x=df['type'].value_counts().reset_index()
x=x.rename(columns={'index':'Type','type':'Count'})
x
```

```python
plt.figure(figsize=[14,4])

# Pie chart
plt.subplot(1,2,1)
plt.pie(x=x['Count'], labels=x['Type'], autopct='%1.1f%%',explode=[0,0.1])
plt.title('Percentage Of Type',fontweight="black",size=20,pad=10)

# Bar plot
plt.subplot(1,2,2)
sns.barplot(data=x, x='Type', y='Count')
plt.title('Count of Each Type',fontweight="black",size=20,pad=10);
```

## Insight 2: Which director has made the maximum number of movies?

```python
x=df['director'].value_counts().head().reset_index()
x
```

According To this df['director'].value_counts() ------>

## Insight 3: In which country have the maximum number of movies been made?

```python
x=df['country'].value_counts().head().reset_index()
x
```

```python
plt.figure(figsize=[15,4])
plt.subplot(1,2,1)
plt.pie(x=x['country'],labels=x['index'],autopct='%0.1f%%',explode=[0,0.1,0.1,0.1,0.1]);
plt.subplot(1,2,2)
sns.barplot(data=x,x=x['index'],y=x['country'])
plt.xlabel('Country')
plt.ylabel('Count')
plt.suptitle('Top 5 Country',size=15,fontweight='black');
```

## Insight 4: Which year has released the highest number of movies or shows?

```python
x=df['release_year'].value_counts().head().reset_index()
x=x.rename(columns={'index':'release_year','release_year':'count'})
x
```

```python
plt.figure(figsize=[15,4])
plt.subplot(1,2,1)
plt.pie(x=x['count'],labels=x['release_year'],autopct='%0.1f%%',explode=[0,0.1,0.1,0.1,0.1])
plt.title('Year Wise Type Percentage',fontweight='black',size=20,pad=10)
plt.subplot(1,2,2)
sns.barplot(data=x,x=x['release_year'],y=x['count'])
plt.title('Year wise type',fontsize=20,fontweight='black');
```

## Insight 5: Highest Rating

```python
x=df['rating'].value_counts().head().reset_index()
x=x.rename(columns={'index':'rating','rating':'count'})
x
```

```python
plt.figure(figsize=(14,4))
plt.subplot(1,2,1)
plt.pie(data=x,x=x['count'],labels='rating',autopct='%0.01f%%',explode=[0,0.1,0.1,0.1,0.1])
plt.title("Rating Percentage",fontweight="black",size=20,pad=10)
plt.subplot(1,2,2)
sns.barplot(data=x,x=x['rating'],y=x['count'])
plt.title("Rating Wise Count",fontweight="black",size=20,pad=10)
```
